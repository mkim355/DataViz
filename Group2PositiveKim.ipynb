```python
import pandas as pd
import numpy as np
import bqplot.pyplot as plt
```

```python
df = pd.read_csv('store_data.csv', encoding='ISO-8859-1')
```

```python
import matplotlib.pyplot as plt

texas_data = df[df['State'] == 'Texas']

category_order = ['Office Supplies', 'Furniture', 'Technology']

# Get the index positions of each category in the desired order
index_positions = [texas_data['Category'].tolist().index(category) for category in category_order]

# Use the index positions to reorder the x and y data
x = category_order
y = texas_data['Sales'].tolist()
y = [y[index] for index in index_positions]

fig = plt.figure(figsize=(10, 6))
fig.suptitle('Sales by Category in Texas')

bar_chart = plt.bar(x, y, color='green')

plt.xlabel('Category', fontsize=16)
plt.ylabel('Sales', fontsize=16)

# Set tick labels and locations manually
bar_chart.scales['x'].tick_values = x
bar_chart.scales['x'].tick_labels = category_order

plt.ylim(0, max(y)*1.1)  # adjust y-axis range

plt.show()
```
